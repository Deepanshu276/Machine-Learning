# Merging The Different DataSet

```python
data=pd.concat([data1,data2,data3,data4,data4])
```

```python
data.shape
```

```python
data
```

```python
#Finding The Sum of null value in each column
data.isnull().sum()
```

```python
#Replacing the Null Value with the mode as these rows cantains less number of nan 
data=data.fillna({'End_Lat':data['End_Lat'].mode()[0]})
data=data.fillna({"Weather_Condition":data['Weather_Condition'].mode()[0]})
data=data.fillna({"Wind_Speed(mph)":data["Wind_Speed(mph)"].mode()[0]})
data=data.fillna({"Wind_Direction":data["Wind_Direction"].mode()[0]})
data=data.fillna({"Visibility(mi)":data["Visibility(mi)"].mode()[0]})
data=data.fillna({"Pressure(in)":data["Pressure(in)"].mode()[0]})
data=data.fillna({"Humidity(%)":data["Humidity(%)"].mode()[0]})
data=data.fillna({"Temperature(F)":data["Temperature(F)"].mode()[0]})
data=data.fillna({"Weather_Timestamp":data["Weather_Timestamp"].mode()[0]})
data=data.fillna({"Airport_Code":data["Airport_Code"].mode()[0]})
data=data.fillna({"Timezone":data["Timezone"].mode()[0]})
data=data.fillna({"Zipcode":data["Zipcode"].mode()[0]})
data=data.fillna({"End_Lng":data["End_Lng"].mode()[0]})
```

```python
data.isnull().sum()
```

```python
#Dropping columns as it contain large number of nan
data_new=data.drop(['Precipitation(in)','Wind_Chill(F)'],inplace=False,axis=1)
```

```python
data_new
```

```python
#Deleting rows which contain nan in number column
data_new=data_new.dropna(subset=['Number'])
```

```python
data_new
```

```python
data_new=data_new.reset_index(drop=True)
```

# Time and date that had reported more accident

```python
data_new
```

```python
#Splitting the Order Date into Date and Time
data_new[['Date', 'Time']] = data_new['Start_Time'].str.split(' ', n=1, expand=True)
```

```python
#converting datatype of Time and extracting hour from it
data_new['Time']= pd.to_datetime(data_new['Time'], errors='coerce')
data_new['hour'] = data_new['Time'].dt.hour
```

```python
import seaborn as sns
sns.set_style("darkgrid")
sns.distplot(data_new.hour, bins = 24, kde=False, norm_hist=True);
```

There are two peaks of time- one in the morning in between 6 am to 9 am and another between 3 pm and 6 pm. This is consistent with the assumption that rush hours in the morning and evening could lead to more accidents.

```python
#Converting datatypes of Start_Time
data_new['Start_Time']= pd.to_datetime(data_new['Start_Time'], errors='coerce')
data_new['Start_Time'] = data_new['Start_Time'].dt.strftime('%Y-%m-%d')
data_new['Start_Time']= pd.to_datetime(data_new['Start_Time'])
```

```python
#Extracting Monday
monday_start_time = data_new.Start_Time[data_new.Start_Time.dt.dayofweek == 0]
```

# Effect of Temperature on Visibility

```python
sns.histplot(data_new['Visibility(mi)'].value_counts(), bins= 5, kde=True, log_scale=True);
```

The majority of accidents happened when the visibility was between 1 to 10 miles. Typically, 10-mile visibility is used as a rule of thumb as a person should be able “to see and identify” in the daytime “a prominent dark object against the sky at the horizon” and at night “a known, preferably unfocused, moderately intense light source” that’s 10 miles away. There’s no definition of how tall the “prominent dark object” or the “light source” can be.
On the surface of things, visibility is not a strong enough reason for the accidents.
